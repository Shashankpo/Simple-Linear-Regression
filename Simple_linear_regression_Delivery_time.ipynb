## Model Building 

```python
import statsmodels.formula.api as smf 
import numpy as np
```

```python
df_cleaned1 = data.rename({'Delivery Time': 'DT'}, axis=1)
df_cleaned1
```

```python
df_cleaned2 = df_cleaned1.rename({'Sorting Time': 'ST'}, axis=1)
df_cleaned2
```

```python
model = smf.ols('DT~ST',data=df_cleaned2).fit()
```

```python
sns.regplot(x='DT',y='ST', data=df_cleaned2);
```

```python
#coefficients
model.params
```

```python
#t and p value
print(model.tvalues,'\n',model.pvalues)
```

```python
#R squared values
(model.rsquared,model.rsquared_adj)
```

## The R^2=0.68 value are so less
## We detect the outlier and remove it
## Delete 20th row

```python
from pandas.io.formats.info import DataFrameTableBuilderNonVerbose
# Discard the data points which are influencers and reassign the row number (reset_index(drop=True))
df=df_cleaned2.drop(df_cleaned2.index[[20]],axis=0).reset_index(drop=True)
df
```

```python
#Correlation
df.corr()
```

```python
model = smf.ols('DT~ST',data=df).fit()
```

```python
sns.regplot(x="ST",y="DT",data=df_cleaned2)  #Scatter Plot
```

```python
#cofficient
model.params
```

```python
#t and p-Values
print(model.tvalues, '\n', model.pvalues)
```

```python
#R squared values
(model.rsquared,model.rsquared_adj)
```
