## Model Building 

```python
import statsmodels.formula.api as smf 
import numpy as np
```

```python
x = data['Delivery Time'].tolist()
y = data['Sorting Time'].tolist()   #Defining the variables
```

```python
sns.regplot(x="Sorting Time",y="Delivery Time",data=data) #scatter plot
```

```python
model = smf.ols("x~y",data = data).fit()
```

```python
# cofficient 
model.params
```

```python
#t and p values
print(model.tvalues, '\n', model.pvalues)
```

```python
#R squared values
(model.rsquared,model.rsquared_adj)
```

## The R^2=0.68 value are so less we decet the outliers and remove it Delete 20th Row
